```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# NASA Mars News Scrape
# Setup Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#URL to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# BeautifulSoup
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Print formatted version of the soup
print(soup.prettify())
```

```python
# Find Latest News Title
news_title = soup.find('div', class_='content_title').text
news_title
```

```python
# Find Latest News Paragraph Text
news_paragraph = soup.find('div', class_='article_teaser_body').text
news_paragraph
```

```python
# Mars Hemispheres
# URL to be scraped
url = 'https://marshemispheres.com/'
browser.visit(url)
```

```python
# BeautifulSoup
url = 'https://marshemispheres.com/'
html = browser.html
soup = bs(html, 'html.parser')
```

```python
description = soup.find_all("div", class_="item")

img_href = []
list_titles = []

for item in description:
    single = item.a

    href = single ['href']
    title = item.h3.text

    img_href.append(href)
    list_titles.append(title)
```

```python
print(list_titles)
```

```python
print(img_href)
```

```python
full_href = []

for i in range(1,5):
    url_titles = {}

    # Copied the xpath of the clickable <h3> to click for full img and loop through for all four
    xpath = '//*[@id="product-section"]/div[2]/div[' + str(i) + ']/div/a/h3'
    browser.find_by_xpath(xpath).click()

    # Saved copied xpath for full_img and title
    full_img = browser.find_by_xpath('//*[@id="wide-image"]/div/ul/li[1]/a')['href']
    img_title = browser.find_by_xpath('//*[@id="results"]/div[1]/div/div[3]/h2').text
    
    # Create dictionaries
    url_titles['img_url'] = full_img
    url_titles['title'] = img_title

    # Append dictionaries to full_href
    full_href.append(url_titles)

    # Return to previous page for next loop
    browser.back()
    
full_href
```

```python
hemisphere_image_urls = {"title":[], 'img_url':[]}
for x in range(4):
    hemisphere_image_urls["title"].append(list_titles[x])
    hemisphere_image_urls["img_url"].append(full_href[x])
```
